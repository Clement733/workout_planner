# Creation of male data

## First male user (name = Robert)

### Chest exercise data

```python
num_values = 100
chest_fm = np.random.randint(np.random.randint(30, 41), np.random.randint(80, 96), num_values)
chest_fm = np.sort(chest_fm)
```

```python
rep_chest_fm = []
rep_count_fm = {}
for repetition in chest_fm:
    if repetition not in rep_count_fm:
        rep_count_fm[repetition] = 8
    else:
        rep_count_fm[repetition] += 2
    rep_chest_fm.append(rep_count_fm[repetition])
df_chest_fm = pd.DataFrame({'Chest_Weight': chest_fm, 'Chest_Repetitions' : rep_chest_fm})
```

### Shoulders exercise data

```python
shoulders_fm = np.random.randint(np.random.randint(6, 8)*2, np.random.randint(11, 13)*2, num_values)/2
shoulders_fm = np.sort(shoulders_fm)
```

```python
rep_shoulders_fm = []
rep_count_s = {}
max_rep = 15
for repetition in shoulders_fm:
    if repetition not in rep_count_fm:
        rep_count_s[repetition] = 8
    else:
        rep_count_s[repetition] += 1
    if rep_count_s[repetition] > max_rep:
        rep_count_s[repetition] = max_rep
    rep_shoulders_fm.append(rep_count_s[repetition])
df_shoulders_fm = pd.DataFrame({'Shoulders_Weight': shoulders_fm, 'Shoulders_Repetitions': rep_shoulders_fm})
```

### Legs exercise data

```python
legs_fm = np.random.randint(60, 121, num_values)
legs_fm = np.sort(legs_fm)
```

```python
rep_legs_fm = []
rep_count = {}
for repetition in legs_fm:
    if repetition not in rep_count:
        rep_count[repetition] = 8
    else:
        rep_count[repetition] += 1
    rep_legs_fm.append(rep_count[repetition])
df_legs_fm = pd.DataFrame({'Legs_Weight': legs_fm, 'Legs_Repetitions': rep_legs_fm})
```

### Back exercise data

```python
back_fm = np.random.randint(20, 81, num_values)
back_fm = np.sort(back_fm)
```

```python
rep_back_fm = []
rep_count = {}
for repetition in back_fm:
    if repetition not in rep_count:
        rep_count[repetition] = 8
    else:
        rep_count[repetition] += 1
    rep_back_fm.append(rep_count[repetition])
df_back_fm = pd.DataFrame({'Back_Weight': back_fm, 'Back_Repetitions': rep_back_fm})
```
