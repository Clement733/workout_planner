# Imports

```python
import pandas as pd
import numpy as np
import csv
```

# Creation of data

```python
np.random.seed(0)
```

## Chest exercise data

```python
num_values = 100
chest = np.random.randint(60, 101, num_values)
chest = np.sort(chest)
```

```python
rep_chest = []
rep_count = {}
for repetition in chest:
    if repetition not in rep_count:
        rep_count[repetition] = 8
    else:
        rep_count[repetition] += 2
    rep_chest.append(rep_count[repetition])
df_chest_improve = pd.DataFrame({'Chest_Weight': chest, 'Chest_Repetitions' : rep_chest})
```

## Shoulders exercise data

```python
shoulders = np.random.randint(6, 16, num_values)
shoulders = np.sort(shoulders)
```

```python
rep_shoulders = []
rep_count_s = {}
max_rep = 15
for repetition in shoulders:
    if repetition not in rep_count:
        rep_count[repetition] = 8
    else:
        rep_count[repetition] += 1
    if rep_count[repetition] > max_rep:
        rep_count[repetition] = max_rep
    rep_shoulders.append(rep_count[repetition])
df_shoulders_improve = pd.DataFrame({'Shoulders_Weight': shoulders, 'Shoulders_Repetitions': rep_shoulders})
```

## Legs exercise data

```python
legs = np.random.randint(60, 121, num_values)
legs = np.sort(legs)
```

```python
rep_legs = []
rep_count = {}
for repetition in legs:
    if repetition not in rep_count:
        rep_count[repetition] = 8
    else:
        rep_count[repetition] += 1
    rep_legs.append(rep_count[repetition])
df_legs_improve = pd.DataFrame({'Legs_Weight': legs, 'Legs_Repetitions': rep_legs})
```

## Back exercise data

```python
back = np.random.randint(20, 81, num_values)
back = np.sort(back)
```

```python
rep_back = []
rep_count = {}
for repetition in legs:
    if repetition not in rep_count:
        rep_count[repetition] = 8
    else:
        rep_count[repetition] += 1
    rep_back.append(rep_count[repetition])
df_back_improve = pd.DataFrame({'Back_Weight': back, 'Back_Repetitions': rep_back})
```

## Creation of full body data

```python
list_robert = [chest, rep_chest, shoulders, rep_shoulders, legs, rep_legs, back, rep_back]
```

```python
df_improve = pd.concat([df_chest_improve, df_back_improve, df_shoulders_improve, df_legs_improve], axis = 1)
df_improve
```

# Data exploration

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 5))
sns.lineplot(data=df_chest_improve, x = 'Chest_Weight', y='Chest_Repetitions', ax = ax1)
ax1.set_title('Chest Improvement', fontsize=15)

sns.lineplot(data=df_shoulders_improve, x='Shoulders_Weight', y='Shoulders_Repetitions', ax = ax2)
ax2.set_title('Shoulders Improvement', fontsize=15)

sns.lineplot(data=df_legs_improve, x='Legs_Weight', y='Legs_Repetitions', ax=ax3)
ax3.set_title('Legs Improvement', fontsize=15)

sns.lineplot(data=df_back_improve, x='Back_Weight', y='Back_Repetitions', ax=ax4)
ax4.set_title('Back Improvement', fontsize=15)

plt.tight_layout()
```

This data was created randomly, but assuming the user progresses steadily, throughout 100 sessions for each muscle group.\
We can see how different the data is depending on the weight chosen, the range of it, and the progressive overload the user does.\
For example, this data would be convenient for a male user. The weight chosen is quite heavy at first, and grows steadily to a point where it can be considered heavy by most gym-goers.

Let's try creating a data for a female user. Still taking random numbers, but accordingly to the gender of the user.

# Creation of female data

## Chest exercise data

```python
chest_f = np.random.randint(20 * 2, 51 * 2, num_values) / 2
chest_f = np.sort(chest_f)
chest_f_r = []
chest_f_r_c = {}
for rep in chest_f:
    if rep not in chest_f_r_c:
        chest_f_r_c[rep] = 8
    else:
        chest_f_r_c[rep] += 2
    chest_f_r.append(chest_f_r_c[rep])
df_chest_f = pd.DataFrame({'Chest_Weight': chest_f, 'Chest_Repetitions': chest_f_r})
```

## Shoulders exercise data

```python
shoulders_f = np.random.randint(4, 11, num_values)
shoulders_f = np.sort(shoulders_f)
shoulders_f_r = []
shoulders_f_r_c = {}
for rep in shoulders_f:
    if rep not in shoulders_f_r_c:
        shoulders_f_r_c[rep] = 8
    else:
        shoulders_f_r_c[rep] += 1
    if shoulders_f_r_c[rep] > 15:
        shoulders_f_r_c[rep] = 15
    shoulders_f_r.append(shoulders_f_r_c[rep])
df_shoulders_f = pd.DataFrame({'Shoulders_Weight': shoulders_f, 'Shoulders_Repetitions': shoulders_f_r})
```

## Legs exercise data

```python
legs_f = np.random.randint(30 * 2, 71 * 2, num_values) / 2
legs_f = np.sort(legs_f)
legs_f_r = []
legs_f_r_c = {}
for rep in legs_f:
    if rep not in legs_f_r_c:
        legs_f_r_c[rep] = 8
    else:
        legs_f_r_c[rep] += 2
    legs_f_r.append(legs_f_r_c[rep])
df_legs_f = pd.DataFrame({'Legs_Weight': legs_f, 'Legs_Repetitions': legs_f_r})
```

## Back exercise data

```python
back_f = np.random.randint(10 * 2, 41 * 2, num_values) / 2
back_f = np.sort(back_f)
back_f_r = []
back_f_r_c = {}
for rep in back_f:
    if rep not in back_f_r_c:
        back_f_r_c[rep] = 8
    else:
        back_f_r_c[rep] += 2
    back_f_r.append(back_f_r_c[rep])
df_back_f = pd.DataFrame({'Back_Weight': back_f, 'Back_Repetitions': back_f_r})
```

## Creation of full body data

```python
list_julia = [chest_f, chest_f_r, shoulders_f, shoulders_f_r, legs_f, legs_f_r, back_f, back_f_r]
```

```python
df_body_f = pd.concat([df_chest_f, df_shoulders_f, df_legs_f, df_back_f], axis = 1)
df_body_f.head()
```

# Data exploration

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12,5))

sns.lineplot(df_chest_f, x='Chest_Weight', y='Chest_Repetitions', ax=ax1)
ax1.set_title('Chest Improvement')

sns.lineplot(df_shoulders_f, x='Shoulders_Weight', y='Shoulders_Repetitions', ax=ax2)
ax2.set_title('Shoulders Improvement')

sns.lineplot(df_legs_f, x='Legs_Weight', y='Legs_Repetitions', ax=ax3)
ax3.set_title('Legs Improvement')

sns.lineplot(df_back_f, x='Back_Weight', y='Back_Repetitions', ax=ax4)
ax4.set_title('Back Improvement')

plt.tight_layout()
```

Although the data is still randomly created, the general consensus is that females are physically weaker than male. This is why I've deliberately chosen lower weights.\
Contrary to its male counterpart, we can see that the female user would stagnate much more on chest and back, and this, due to the lower range of weights. On the contrary, males tend to stagnate more for the legs. Thus, the model should take this into account too.
