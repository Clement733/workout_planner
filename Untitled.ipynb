```python
import pandas as pd
import numpy as np
import csv
```

```python
np.random.seed(0)
```

```python
num_values = 100
chest = np.random.randint(60, 101, num_values)
chest = np.sort(chest)
```

```python
shoulders = np.random.randint(6, 16, num_values)
shoulders = np.sort(shoulders)
```

```python
legs = np.random.randint(60, 121, num_values)
legs = np.sort(legs)
```

```python
back = np.random.randint(20, 81, num_values)
back = np.sort(back)
```

```python
rep_chest = []
rep_count = {}
for repetition in chest:
    if repetition not in rep_count:
        rep_count[repetition] = 8
    else:
        rep_count[repetition] += 2
    rep_chest.append(rep_count[repetition])
rep_chest
```

```python
df_chest_improve = pd.DataFrame({'Chest_Weight': chest, 'Repetitions' : rep_chest})
df_chest_improve.head()
```

```python
rep_shoulders = []
rep_count = {}
max_rep = 15
for repetition in shoulders:
    if repetition not in rep_count:
        rep_count[repetition] = 8
    else:
        rep_count[repetition] += 1
    if rep_count[repetition] > max_rep:
        rep_count[repetition] = max_rep
    rep_shoulders.append(rep_count[repetition])
rep_shoulders
```

```python
df_shoulders_improve = pd.DataFrame({'Shoulders_Weight': shoulders, 'Repetitions': rep_shoulders})
df_shoulders_improve
```

```python
rep_legs = []
rep_count = {}
for repetition in legs:
    if repetition not in rep_count:
        rep_count[repetition] = 8
    else:
        rep_count[repetition] += 1
    rep_legs.append(rep_count[repetition])
rep_legs
```

```python
df_legs_improve = pd.DataFrame({'Legs_Weight': legs, 'Repetitions': rep_legs})
df_legs_improve
```

```python
rep_back = []
rep_count = {}
for repetition in legs:
    if repetition not in rep_count:
        rep_count[repetition] = 8
    else:
        rep_count[repetition] += 1
    rep_back.append(rep_count[repetition])
rep_back
```

```python
df_back_improve = pd.DataFrame({'Back_Weight': back, 'Repetitions': rep_back})
df_back_improve
```

```python
df_improve = pd.concat([df_chest_improve, df_back_improve, df_shoulders_improve, df_legs_improve], axis = 1)
df_improve
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 5))
sns.lineplot(data=df_chest_improve, x = 'Chest_Weight', y='Repetitions', ax = ax1)
ax1.set_title('Chest Improvement', fontsize=15)

sns.lineplot(data=df_shoulders_improve, x='Shoulders_Weight', y='Repetitions', ax = ax2)
ax2.set_title('Shoulders Improvement', fontsize=15)

sns.lineplot(data=df_legs_improve, x='Legs_Weight', y='Repetitions', ax=ax3)
ax3.set_title('Legs Improvement', fontsize=15)

sns.lineplot(data=df_back_improve, x='Back_Weight', y='Repetitions', ax=ax4)
ax4.set_title('Back Improvement', fontsize=15)

plt.tight_layout()
```
